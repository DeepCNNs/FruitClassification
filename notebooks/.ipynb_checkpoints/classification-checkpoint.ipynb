```python
# TensorFlow and tf.keras
import tensorflow as tf
from tensorflow.keras import datasets, layers, models

# Helper libraries
import numpy as np
import matplotlib.pyplot as plt
import enum

import cv2
from os import listdir
from os.path import isfile, join

print(tf.__version__)
```

```python
train_images_dir = "/home/adity/Desktop/projects/datasets/fruits/final_train/ori/"

def read_files_from_dir(mypath):
    allfiles = [f for f in listdir(mypath) if isfile(join(mypath, f))]
    return allfiles

train_images_files = read_files_from_dir(train_images_dir)
train_images_list = []
train_labels_list = []

for file in train_images_files:
    filename = file.split('.')[0]
    img_class = filename.split('_')[0]
    image = cv2.imread(train_images_dir + file)
#     image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    train_images_list.append(image)
    
    if img_class == 'apple':
        train_labels_list.append(0)
    elif img_class == 'banana':
        train_labels_list.append(1)
    else: #orange
        train_labels_list.append(2)

train_labels = np.array(train_labels_list)
train_images = np.array(train_images_list)
train_images.shape
train_labels

test_images_dir = "/home/adity/Desktop/projects/datasets/fruits/final_test/ori/"
test_images_files = read_files_from_dir(test_images_dir)
test_images_list = []
test_labels_list = []

for file in test_images_files:
    filename = file.split('.')[0]
    img_class = filename.split('_')[0]
    image = cv2.imread(test_images_dir + file)
#     image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    test_images_list.append(image)
    
    if img_class == 'apple':
        test_labels_list.append(0)
    elif img_class == 'banana':
        test_labels_list.append(1)
    else: #orange
        test_labels_list.append(2)

test_labels = np.array(test_labels_list)
test_images = np.array(test_images_list)
test_images.shape
test_labels
```

```python

# Normalize pixel values to be between 0 and 1
train_images, test_images = train_images / 255.0, test_images / 255.0
```

```python
plt.figure(figsize=(10,10))
for i in range(25):
    plt.subplot(5,5,i+1)
    plt.xticks([])
    plt.yticks([])
    plt.grid(False)
    plt.imshow(train_images[i], cmap=plt.cm.binary)
    # The CIFAR labels happen to be arrays, 
    # which is why you need the extra index
    plt.xlabel(class_names[train_labels[i]])
plt.show()
```

```python
model = models.Sequential()
model.add(layers.Conv2D(28, (3, 3), activation='relu', input_shape=(28, 28, 3)))
model.add(layers.MaxPooling2D((2, 2)))
model.add(layers.Conv2D(56, (3, 3), activation='relu'))
model.add(layers.MaxPooling2D((2, 2)))
model.add(layers.Conv2D(56, (3, 3), activation='relu'))
```

```python
model.summary()
```

```python
model.add(layers.Flatten())
model.add(layers.Dense(56, activation='relu'))
model.add(layers.Dense(3))
```

```python
model.summary()
```

```python
# model.compile(optimizer='adam',
#               loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True),
#               metrics=['accuracy'])

model.compile(optimizer=tf.keras.optimizers.Adam(0.0001),
              loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True),
              metrics=['accuracy'])

history = model.fit(train_images, train_labels, epochs=10, 
                    validation_data=(test_images, test_labels))
```

```python
plt.plot(history.history['acc'], label='accuracy')
plt.plot(history.history['val_acc'], label = 'val_accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.ylim([0.5, 1])
plt.legend(loc='lower right')

test_loss, test_acc = model.evaluate(test_images,  test_labels, verbose=2)
```

```python
def create_model():
    model = models.Sequential()
    model.add(layers.Conv2D(28, (3, 3), activation='relu', input_shape=(28, 28, 3)))
    model.add(layers.MaxPooling2D((2, 2)))
    model.add(layers.Conv2D(56, (3, 3), activation='relu'))
    model.add(layers.MaxPooling2D((2, 2)))
    model.add(layers.Conv2D(56, (3, 3), activation='relu'))
    return model


# load json and create model
loaded_model = models.load_weights('/home/adity/Desktop/projects/assignment/model/fruits.h5')

```
